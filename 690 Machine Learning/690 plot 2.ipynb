```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
beer= pd.read_csv(r'C:\Users\ywan3\Desktop\mvbeer.csv',nrows=3000000)
beer
```

```python
beer.type.unique()
```

```python
import seaborn as sns
saleB= beer.loc[beer["sale_B"]==1]
g = sns.countplot(x="upc",data=saleB)
g.axes.get_xaxis().set_ticks([])
```

```python
g = sns.countplot(x="store_id",data=saleB)
g.axes.get_xaxis().set_ticks([])
```

```python
saleS= beer.loc[beer["sale_S"]==1]
g = sns.countplot(x="upc",data=saleS)
g.axes.get_xaxis().set_ticks([])
```

```python
g = sns.countplot(x="store_id",data=saleS)
g.axes.get_xaxis().set_ticks([])
```

```python
saleC= beer.loc[beer["sale_C"]==1]
g = sns.countplot(x="upc",data=saleC)
g = plt.setp(g.get_xticklabels(), rotation=45) 
plt.locator_params(axis='x', nbins=20)
```

```python
g = sns.countplot(x="store_id",data=saleC)
g.axes.get_xaxis().set_ticks([])
```

```python
sale= beer.loc[beer["sale"]==1]
s = pd.Series(sale['upc'], dtype="category")
g = sns.countplot(x=s,data=sale)
g.axes.get_xaxis().set_ticks([])
```
